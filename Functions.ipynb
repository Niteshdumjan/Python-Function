```python
# 5. What are iterators in Python and how do they differ from iterables?
'''
  iterators and iterables are closely related but have distinct differences.
'''
# 1. What is an Iterable?
'''
An iterable is any Python object that can return an iterator.
It is an object that implements the __iter__() method, which returns an iterator.
'''

# Examples of Iterables:
  # Lists: [1, 2, 3]

  # Tuples: (4, 5, 6)

  # Strings: "hello"

  # Dictionaries: {"a": 1, "b": 2}

  # Sets: {7, 8, 9}

  # Ranges: range(10)

numbers = [1, 2, 3]
iterator = iter(numbers)

print(next(iterator))
print(next(iterator))
print(next(iterator))

# 2. What is an Iterator?

'''
An iterator is an object that produces values one at a time when iterated over.
It must implement two methods:
'''
# __iter__(): Returns the iterator object itself.

# __next__(): Returns the next item in the sequence.

# Example

class Counter:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.end:
            raise StopIteration
        self.current += 1
        return self.current - 1

counter = Counter(1, 4)
for num in counter:
    print(num)
```

```python
# 6. Explain the concept of generators in Python and how they are defined.
'''
A generator in Python is a special type of iterator that generates values lazily,
meaning it produces one item at a time instead of storing everything in memory.
Generators are created using the yield keyword.
'''

# Generators can be created in two ways:

  # 1. Using a generator function (yield)

  # 2. Using a generator expression ((...))

# Generator Function (yield keyword)
# Code
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

gen = count_up_to(3)

print(next(gen))
print(next(gen))
print(next(gen))

# Generator Expression ((...))
# Example
gen_exp = (x * x for x in range(5))

print(next(gen_exp))
print(next(gen_exp))
print(next(gen_exp))
```

```python
# 7. What are the Advantages of Using Generators Over Regular Functions in Python
'''
  Generators offer several benefits over regular functions, especially when
  dealing with large datasets or infinite sequences. Below are the key advantages:
'''
# 1. Memory Efficiency
'''Generators do not store all values in memory; they generate values on demand.'''

# 2. Faster Execution (Lazy Evaluation)
'''
Generators execute lazily, computing values only when needed.
A regular function processes everything at once, which can be slow.
'''

# 3. Supports Infinite Sequences
'''
Generators can produce infinite sequences, unlike lists which must be finite.
A regular function must return all values at once, but a generator can keep
producing values indefinitely.
'''

# 4. Better Performance for Large Files (Streaming Data)
'''Generators are perfect for reading large files line by line, avoiding memory overload.'''
```

```python
# 8. What is a lambda function in Python and when is it typically used?
# Lambda Function in Python
'''
  A lambda function in Python is a small, anonymous function that can have any number
  of arguments but only one expression. It is defined using the lambda keyword instead of def.
'''
# 3. When to Use Lambda Functions?
# 1. Short and Simple Operations
'''Lambda functions are used for quick, one-line operations where defining a full function is unnecessary.'''


# Example
# Regular Function
def add(x, y):
    return x + y

# Equivalent Lambda Function
add_lambda = lambda x, y: x + y


print(add(3, 5))
print(add_lambda(3, 5))

```

```python
# 9. Explain the purpose and usage of the `map()` function in Python.

# map() Function in Python
'''
  The map() function in Python is used to apply a given function to each item in
  an iterable (like a list, tuple, or set) and return a new iterable (map object).
 '''

#  Example: Using map() with a Regular Function

def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

# Convert map object to a list
print(list(squared_numbers))

# Using map() with a Lambda Function
''' Lambda functions are often used with map() for simplicity. '''

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x * x, numbers)
print(list(squared_numbers))


# Using map() with Multiple Iterables
'''
  map() can process multiple iterables in parallel if the function takes multiple
  arguments.
'''

a = [1, 2, 3]
b = [4, 5, 6]

sum_ab = map(lambda x, y: x + y, a, b)
print(list(sum_ab));


# Using map() to Convert Data Types.
nums = ["1", "2", "3"]
nums_int = map(int, nums)
print(list(nums_int))



```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
# Difference Between map(), reduce(), and filter() in Python

'''
  These three functions—map(), reduce(), and filter()—are part of Python’s
  functional programming tools. They allow operations on iterables without explicit loops.
'''

# 1. map() - Apply a funtion to every Element
'''
  Purpose: Transforms each element in an iterable using a function.
  Returns: A map object (convertible to a list, tuple, etc.).
'''
# Example: Squaring Numbers

numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)

print(list(squared))

# 2. filter() – Select Elements That Meet a Condition
'''
  Purpose: Filters elements based on a function that returns True or False.
  Returns: A filter object (convertible to a list, tuple, etc.).
'''

# Example: Filtering Even Numbers
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)

print(list(evens))


# 3. reduce() – Apply a Function Cumulatively
'''
  Purpose: Reduces an iterable to a single value using a cumulative function.
  Returns: A single result instead of a collection.
  Requires: Importing from functools.

'''
# Example: Multiplying All Numbers
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)

print(product)
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

# Code
def reverse_string_while(s):

    reversed_str = ""
    index = len(s) - 1
    while index >= 0:
        reversed_str += s[index]
        index -= 1
    return reversed_str

text = "Niteh dumjan"
print(reverse_string_while(text))

```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

def fib(n):
  a = 0
  b = 1
  if n == 1:
    print(a)

  elif n < 0:
    print("Invalid Number")
  else:
    print(a)
    print(b)

  for i in range(2, n):
    c = a+b
    a = b
    b = c
    print(c)

fib(10)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(max_exponent):
  for exponent in range(max_exponent + 1):
    yield 2 ** exponent

for power in power_of_two(5):
  print(power)
```

```python
# 9. Explain the purpose and usage of the `map()` function in Python.

# map() Function in Python
'''
  The map() function in Python is used to apply a function to every item in an
  iterable (like a list or tuple) and return a new iterable (a map object, which is an iterator). It allows for efficient element-wise transformations without needing an explicit loop.
'''
# Syntax
# map(function, iterable)

# function: A function that takes one or more arguments and returns a result.
# iterable: An iterable (like a list, tuple, or string) that supplies the elements for the function.

# Examples of map() Usage
def square(x):
  return x ** 2

numbers = [1, 2, 3, 4]
squared_numbers = list(map(square, numbers))
print(squared_numbers)

```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
  vowels = "aeiouAEIOU"
  filter_char = filter(lambda char: char not in vowels, string)
  return "".join(filter_char)

input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)
```
